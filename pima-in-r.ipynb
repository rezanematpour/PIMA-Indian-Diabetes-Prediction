```r
d$diastolic[d$diastolic==0]=NA
d$glucose[d$glucose==0]=NA
d$triceps[d$triceps==0]=NA
d$bmi[d$bmi==0]=NA
d$insulin[d$insulin==0]=NA
summary(d)
```

**Missing Values in Observations**

* **Define a function to give the missing values percentage of observations**
* **Functions you may need: is.na(),sum, length(), apply**

```r
missnum<-function(x){a<-sum(is.na(x))/length(x); return(a)}
miss<-apply(d,1,missnum)
```

**Impute the Missing Values**

```r
impu <- mice(d,m=5,maxit=50,meth='pmm')
```

* d: is your data set
* m: is the number of imputed datasets
* maxit: the maximum number of iterations in the algorithm
* meth: the method that you want to use for imputation(predictive mean matching). methods(mice)

**Check the Imputed Data**

```r
impu$imp$bmi
```

**Replace the Imputed Data**

```r
com<- complete(impu,1)
summary(com)
```

**Performance Analysis**

```r
xyplot(impu, insulin ~ triceps+glucose+diastolic| .imp, pch = 20, cex = 1.4)
stripplot(impu, pch = 20, cex = 1.2)
densityplot(impu)
```
